```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel("data.xlsx")
```

```python
data = data[['Price [USD2018/kWh]','NMC Cumulative Demand [GWh]']][:13]
```

```python
#scatter plot
x = data['NMC Cumulative Demand [GWh]']
y = data['Price [USD2018/kWh]']
plt.xlabel('NMC Cumulative Demand [GWh]')
plt.ylabel('Price [USD2018/kWh]')
plt.scatter(x, y)
plt.show()
```

```python
# log transformation on x
data['logx'] = [np.log(x) for x in list(data['NMC Cumulative Demand [GWh]'])]
```

```python
#scatter plot between price and logx

x = data['logx']
y = data['Price [USD2018/kWh]']
plt.xlabel('log NMC Cumulative Demand [GWh]')
plt.ylabel('Price [USD2018/kWh]')
plt.scatter(x, y)
plt.show()
```

```python
# log transformation on y
data['logy'] = [np.log(x) for x in list(data['Price [USD2018/kWh]'])]
```

```python
#scatter plot between logy and NMC Cumulative Demand [GWh](x)

x = data['NMC Cumulative Demand [GWh]']
y = data['logy']
plt.xlabel('NMC Cumulative Demand [GWh]')
plt.ylabel('log Price [USD2018/kWh]')
plt.scatter(x, y)
plt.show()
```

```python
#scatter plot between logy and log(x)
x = data['logx']
y = data['logy']
plt.xlabel('log NMC Cumulative Demand [GWh]')
plt.ylabel('log Price [USD2018/kWh]')
plt.scatter(x, y)
plt.show()
```
